```python
well = np.random.choice(plate1_well_map["Mix1"])
image_path = np.random.choice(experiment_acquisition[experiment_acquisition.well_loc == well].file_path)

# read in the image
image = utils.database_utils.read_image(image_path)

#create composite image
bb_comp = utils.image_utils.create_composite_brainbow_image(image, channel_map, True)

fig, ax = plt.subplots(1,1, dpi = 300, figsize=(3,3))

# visualise image
img_t = np.stack([exposure.equalize_adapthist(bb_comp[:,:,0], clip_limit=0.3),
                  exposure.equalize_adapthist(bb_comp[:,:,1], clip_limit=0.3),
                  exposure.equalize_adapthist(bb_comp[:,:,2], clip_limit=0.3)],axis = 2)
plt.imshow(exposure.adjust_gamma(img_t,gamma = 8, gain =10))
plt.axis('off')
```
