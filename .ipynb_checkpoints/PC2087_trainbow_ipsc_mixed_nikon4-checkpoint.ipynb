```python
import numpy as np
import cv2 as cv

def normalize(raw_image:np.ndarray):
    image = cv.normalize(
        raw_image, None, alpha=0, beta=255, norm_type=cv.NORM_MINMAX, dtype=cv.CV_32F
    )
    return image
```
